**Convert Price Columns into Numeric**

```python
#check the price columns
df.columns
```

```python
df[['initial_price','final_price']].head(10)
```

```python
price_columns = ['initial_price', 'final_price']

for col in price_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('"', '', regex=False)   # Remove quotes
        .str.replace('₹', '', regex=False)   # Remove rupee symbol
        .str.replace(',', '', regex=False)   # Remove commas
        .str.strip()
    )

    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#check the price columns after cinverting to numeric
df.dtypes
```

```python
df[["initial_price","final_price"]].head()
```

**Handle The Missing Values**

```python
#Identify and count the missing values in each cloumn 
df.isnull().sum()
```

```python
#Identify and count the missing values in each cloumn in percentage
(df.isnull().sum()/len(df))*100
```

**Fill Missing values**

**For Numericals Columns**

```python
#Fill missing values for numerical columns with mean
numerical_columns = df.select_dtypes(include=[np.number]).columns

for col in numerical_columns:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
#After filling the missing values in Numerical columns, checking the missing values again
df.isnull().sum()
```

**For Categorical Column**

```python
#Checking the missing values in Categorical columns
df.isnull().sum()
```

```python
#Fill missing values for categorical columns with mode
categorical_columns = df.select_dtypes(include=[object]).columns

for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#After filling the missing values in Categorical columns, checking the missing values again
df.isnull().sum()
```

**Drop Missing Values**

```python
#Checking the missing values in the entire dataset 
df.isnull().sum()
```

```python
#Drop the rows with missing values in the entire dataset
df.dropna(inplace=True)
```

```python
#After dropping the rows with missing values, checking the missing values again
df.isnull().sum()
```

**Univariate Analysis**

```python
#Price distribution
#Histogram

plt.figure(figsize=(10,6))
sns.histplot(df['final_price'], bins=30)

plt.title('Distribution of Product Price')
plt.xlabel('Final Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Rating Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['rating'], bins=30)
plt.title("Product Rating Distribution")
plt.show()
```

**Bivariate Analysis**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='rating',
    y='final_price'
)

plt.title("Rating vs Price")

plt.show()
```

**Category Level Analysis**

```python
#check Category
df['category'].value_counts()
```

```python
#Product per Category
plt.figure(figsize=(12,6))

df['category'].value_counts().plot(
    kind='bar'
)

plt.title("Product by Category")

plt.xlabel("Category")
plt.ylabel("Number of Products")

plt.show()
```

**Average Price by Category**

```python
df['final_price'] = pd.to_numeric(
    df['final_price'].astype(str).str.replace(r'[₹",]', '', regex=True),
    errors='coerce'
)

category_price = df.groupby('category')['final_price'].mean().sort_values()

category_price
```

```python
plt.figure(figsize=(10,6))

category_price.plot(
    kind='bar'
)

plt.title("Average Price by Category")

plt.ylabel("Average Price")

plt.show()
```

**Boxplot Analysis**

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    y=df["final_price"]
)

plt.title("Price Outliers detection")

plt.show()
```
